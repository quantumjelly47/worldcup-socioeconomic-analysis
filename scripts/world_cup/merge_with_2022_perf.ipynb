```python
import pandas as pd
import numpy as np
```

```python
# get 2022 matches
archive_matches_df = pd.read_csv('../../data/raw_data_files/World Cup Data/archive/matches_1930_2022.csv')
matches_2022 = archive_matches_df[archive_matches_df['Year']>=2022]
matches_2022.head()
```

```python
matches_2022.columns
```

```python
# get relevant columns, along with scores for shootouts
matches_2022 = matches_2022[['home_team', 'away_team', 'home_score', 'away_score', 'Round', 'Date', 
'Score', 'Host', 'Year', 'home_penalty_shootout_goal_long', 'away_penalty_shootout_goal_long']]\
    .assign(home_penalty_shootout_goal_long=lambda x: x['home_penalty_shootout_goal_long'].str.split(',').str.len(), 
            away_penalty_shootout_goal_long=lambda x: x['away_penalty_shootout_goal_long'].str.split(',').str.len())\
                .rename(columns={'home_penalty_shootout_goal_long':'home_shootout', 
                                 'away_penalty_shootout_goal_long':'away_shootout'})

matches_2022 = matches_2022.rename(columns={'Round':'Stage', 'Host':'Host Team', 'home_team':'Home Team', 
                                            'away_team':'Away Team', 'home_score':'Home Goals', 'away_score':'Away Goals'})

matches_2022
```

```python
# reformat data into team, away/home, opponent, stage, date, goals_scored, goals_conceded, shootout_status
home_df = matches_2022.assign(
    team=matches_2022['Home Team'],
    home_away='home',
    opponent=matches_2022['Away Team'],
    goals_scored=matches_2022['Home Goals'],
    goals_conceded=matches_2022['Away Goals'],
    shootout_status=lambda d: np.where(
        d['home_shootout'].isna() & d['away_shootout'].isna(),
        np.nan,
        (d['home_shootout'].fillna(0) > d['away_shootout'].fillna(0)).astype(int)
    )
)[['team', 'home_away', 'opponent', 'Stage', 'Date', 'goals_scored', 'goals_conceded', 'shootout_status']]

away_df = matches_2022.assign(
    team=matches_2022['Away Team'],
    home_away='away',
    opponent=matches_2022['Home Team'],
    goals_scored=matches_2022['Away Goals'],
    goals_conceded=matches_2022['Home Goals'],
    shootout_status=lambda d: np.where(
        d['home_shootout'].isna() & d['away_shootout'].isna(),
        np.nan,
        (d['away_shootout'].fillna(0) > d['home_shootout'].fillna(0)).astype(int)
    )
)[['team', 'home_away', 'opponent', 'Stage', 'Date', 'goals_scored', 'goals_conceded', 'shootout_status']]

reshaped_df = pd.concat([home_df, away_df], ignore_index=True).reset_index(drop=True)
reshaped_df.head()
```

```python
# get outcome of match
reshaped_df['win'] = np.where(
    (reshaped_df['goals_scored'] > reshaped_df['goals_conceded']) | (reshaped_df['shootout_status'] == 1),
    1,
    np.where(
        (reshaped_df['goals_scored'] < reshaped_df['goals_conceded']) | (reshaped_df['shootout_status'] == 0),
        0,
        0  # draw but no shootout -> 0
    )
)

reshaped_df.head()
```

```python
# get max stage reached
stage_order = {'Group stage':1, 'Round of 16':2, 'Quarter-finals':3}
reshaped_df['stage_numeric'] = np.where(
    (reshaped_df['Stage']=='Final') & (reshaped_df['win']==1),
    7,
    np.where(
        (reshaped_df['Stage']=='Final') & (reshaped_df['win']==0),
        6,
        np.where(
            (reshaped_df['Stage']=='Third-place match') & (reshaped_df['win']==1),
            5,
            np.where(
                (reshaped_df['Stage']=='Third-place match') & (reshaped_df['win']==0),
                4,
                reshaped_df['Stage'].map(stage_order)
            )
        )
    )
)
reshaped_df.head()
```

```python
# summarize values by team
df_2022 = reshaped_df.groupby('team').agg(
    max_stage_numeric=('stage_numeric', 'max'),
    matches_played=('opponent', 'count'),
    matches_won=('win', 'sum'),
    goals_for=('goals_scored', 'sum'),
    goals_against=('goals_conceded', 'sum')
).reset_index()
```

```python
 # merge with original df to get back win status
df_2022 = df_2022.merge(reshaped_df[['team', 'stage_numeric', 'win']], left_on=['team', 'max_stage_numeric'], right_on=['team', 'stage_numeric'])
df_2022.head()
```

```python
# retrieve max_stage_reached in words
df_2022['max_stage'] = np.where(
    df_2022['max_stage_numeric']==7, 
    'Winner',
    np.where(df_2022['max_stage_numeric']==6,
    'Runners-up',
    np.where(df_2022['max_stage_numeric']==5,
    'Third place',
    np.where(df_2022['max_stage_numeric']==4,
    'Fourth place',
    np.where(df_2022['max_stage_numeric']==3,
    'Quarter-finals',
    np.where(df_2022['max_stage_numeric']==2,
    'Round of 16',
    'Group stage'))))))

df_2022 = df_2022.drop(columns=['win', 'stage_numeric']).drop_duplicates().reset_index(drop=True)
```

```python
def preprocess_year_data(year):
    """
        Processes the World Cup match data for a given year and summarizes, for each team, the max stage of World Cup reached,
        the total number of matches played, the total number of matches won, the total number of goals for and against the team.

        Args:
            year (int): Year of the World Cup of interest.
        
        Returns:
            df_by_year (pd.DataFrame): DataFrame containing summarized data for every qualified team in a given World Cup year,
                including the max stage of World Cup reached, the total number of matches played, the total number of matches won, 
                the total number of goals for and against the team.
    """
    # get year's matches
    archive_matches_df = pd.read_csv('../../data/raw_data_files/World Cup Data/archive/matches_1930_2022.csv')
    matches = archive_matches_df[archive_matches_df['Year']==year]

    # get relevant columns, along with scores for shootouts
    matches = matches[['home_team', 'away_team', 'home_score', 'away_score', 'Round', 'Date', 
    'Score', 'Host', 'Year', 'home_penalty_shootout_goal_long', 'away_penalty_shootout_goal_long']]\
        .assign(home_penalty_shootout_goal_long=lambda x: x['home_penalty_shootout_goal_long'].str.split(',').str.len(), 
                away_penalty_shootout_goal_long=lambda x: x['away_penalty_shootout_goal_long'].str.split(',').str.len())\
                    .rename(columns={'home_penalty_shootout_goal_long':'home_shootout', 
                                    'away_penalty_shootout_goal_long':'away_shootout'})
    # rename columns
    matches = matches.rename(columns={'Round':'Stage', 'Host':'Host Team', 'home_team':'Home Team', 
                                                'away_team':'Away Team', 'home_score':'Home Goals', 'away_score':'Away Goals'})

    # reformat data into team, away/home, opponent, stage, date, goals_scored, goals_conceded, shootout_status
    home_df = matches.assign(
        team=matches['Home Team'],
        home_away='home',
        opponent=matches['Away Team'],
        goals_scored=matches['Home Goals'],
        goals_conceded=matches['Away Goals'],
        shootout_status=lambda d: np.where(
            d['home_shootout'].isna() & d['away_shootout'].isna(),
            np.nan,
            (d['home_shootout'].fillna(0) > d['away_shootout'].fillna(0)).astype(int)
        )
    )[['team', 'home_away', 'opponent', 'Stage', 'Date', 'goals_scored', 'goals_conceded', 'shootout_status']]

    away_df = matches.assign(
        team=matches['Away Team'],
        home_away='away',
        opponent=matches['Home Team'],
        goals_scored=matches['Away Goals'],
        goals_conceded=matches['Home Goals'],
        shootout_status=lambda d: np.where(
            d['home_shootout'].isna() & d['away_shootout'].isna(),
            np.nan,
            (d['away_shootout'].fillna(0) > d['home_shootout'].fillna(0)).astype(int)
        )
    )[['team', 'home_away', 'opponent', 'Stage', 'Date', 'goals_scored', 'goals_conceded', 'shootout_status']]

    reshaped_df = pd.concat([home_df, away_df], ignore_index=True).reset_index(drop=True)

    # get outcome of match
    reshaped_df['win'] = np.where(
        (reshaped_df['goals_scored'] > reshaped_df['goals_conceded']) | (reshaped_df['shootout_status'] == 1),
        1,
        np.where(
            (reshaped_df['goals_scored'] < reshaped_df['goals_conceded']) | (reshaped_df['shootout_status'] == 0),
            0,
            0  # draw but no shootout -> 0
        )
    )

    # get max stage reached
    stage_order = {'Group stage':1, 'Round of 16':2, 'Quarter-finals':3}
    reshaped_df['stage_numeric'] = np.where(
        (reshaped_df['Stage']=='Final') & (reshaped_df['win']==1),
        7,
        np.where(
            (reshaped_df['Stage']=='Final') & (reshaped_df['win']==0),
            6,
            np.where(
                (reshaped_df['Stage']=='Third-place match') & (reshaped_df['win']==1),
                5,
                np.where(
                    (reshaped_df['Stage']=='Third-place match') & (reshaped_df['win']==0),
                    4,
                    reshaped_df['Stage'].map(stage_order)
                )
            )
        )
    )

    # summarize values by team
    df_by_year = reshaped_df.groupby('team').agg(
        max_stage_numeric=('stage_numeric', 'max'),
        matches_played=('opponent', 'count'),
        matches_won=('win', 'sum'),
        goals_for=('goals_scored', 'sum'),
        goals_against=('goals_conceded', 'sum')
    ).reset_index()

    # merge with original df to get back win status
    df_by_year = df_by_year.merge(
        reshaped_df[['team', 'stage_numeric', 'win']], left_on=['team', 'max_stage_numeric'], right_on=['team', 'stage_numeric']
    )

    # retrieve max_stage_reached in words
    df_by_year['max_stage'] = np.where(
        df_by_year['max_stage_numeric']==7, 
        'Winner',
        np.where(df_by_year['max_stage_numeric']==6,
        'Runners-up',
        np.where(df_by_year['max_stage_numeric']==5,
        'Third place',
        np.where(df_by_year['max_stage_numeric']==4,
        'Fourth place',
        np.where(df_by_year['max_stage_numeric']==3,
        'Quarter-finals',
        np.where(df_by_year['max_stage_numeric']==2,
        'Round of 16',
        'Group stage'))))))

    df_by_year = df_by_year.drop(columns=['win', 'stage_numeric']).drop_duplicates().reset_index(drop=True)
    
    return df_by_year
```
